&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;En el siguiente trabajo realizaremos un análisis sobre cómo fue el desarrollo de la materia a lo largo del año.
Para esto creamos un formulario (https://forms.gle/hPMtWem3RdoEWjFJ7) en el que votaron todos los integrantes del curso. Luego de eso,
generamos un csv para trabajar los datos desde ahí.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Para empezar, podemos ver un gran porcentaje de alumnos que no tienen todos los trabajos entregados al día, siendo solo
el 37% quienes pudieron cumplir con las fechas establecidas.

```python
import pandas as pd
import matplotlib.pyplot as plt1
import numpy as np

df1 = pd.read_csv("../data/analisisTaller.csv",sep=",",engine="python" )
```

```python
#grafico si/no torta
plt1.title('Porcentaje de alumnos que entregaron todos los trabajos')
cantidad = [df1["¿Pudo entregar todos los trabajos de la materia?"].value_counts()["Sí"],df1["¿Pudo entregar todos los trabajos de la materia?"].value_counts()["No"]]
atributos = ["Sí","No"]
plt1.pie(cantidad, labels=atributos, autopct='%1.1f%%')
plt1.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Frente a este alto porcentaje de estudiantes que adeudan trabajos, decidimos analizar a qué se debe su bajo rendimiento en la materia.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Uno de los motivos podría ser su interés sobre el lenguaje que se utiliza en la materia (Python). Sin embargo, una gran parte del
alumnado se siente conforme con el lenguaje y, además, no le parece aburrido el análisis de datos.

```python
#grafico python
x_values = df1['¿Cuánto te gusta Python como lenguaje de programación?'].unique()
y_values = df1['¿Cuánto te gusta Python como lenguaje de programación?'].value_counts().tolist()
y_values.reverse()
x_values= np.flip(x_values)
# gráfico meses
plt1.bar(x_values,y_values)
plt1.xlabel('Valoraciones')
plt1.ylabel('Alumnos')
plt1.xticks(rotation=63,fontsize=11)
plt1.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Desde el año pasado que el sistema educativo se encuentra en constante cambio debido al Coronavirus. Esto hizo que durante el transcurso del 2021 la modalidad de enseñanza se vea alterada
oscilando entre lo virtual y lo presencial. Esto repercute en la organización de los estudiantes y, en consecuencia, su cumplimiento en las entregas.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Sin embargo, si bien en agosto se volvió a la presencialidad completa (forma de enseñanaza que prefieren la mayoría de los encuestados),
en este momento es cuando se puede ver que dejan de entregar los trabajos.

```python
x_values = df1['¿Hasta qué mes entregaste trabajos?'].unique()
y_values = df1['¿Hasta qué mes entregaste trabajos?'].value_counts().tolist()
y_values.reverse()
x_values= np.flip(x_values)
# gráfico meses
plt1.bar(x_values,y_values)
plt1.xlabel('Meses')
plt1.ylabel('Alumnos')
plt1.xticks(rotation=63,fontsize=11)
plt1.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A pesar de que más del 50% ha tenido alguna dificultad para la búsqueda del dataset, ésto no fue un motivo de retraso ya que la mayoría de los trabajos que
fueron dados a partir de septiembre tenían los datasets involucrados o, en su defecto, no eran necesarios y había que buscar otras cosas (Páginas web, APIs).
